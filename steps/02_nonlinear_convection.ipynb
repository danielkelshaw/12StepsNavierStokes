## **1D Non-Linear Convection:**

The 1-D Convection equation is:

$$  \frac{\partial{}u}{\partial{t}} + c\frac{\partial{u}}{\partial{x}} = 0  $$

Instead of a constant, $c$, we use the solution $u$ to multiply the second term.

We can discretise the non-linear convection equation as follows:

$$  \frac{u_{i}^{n + 1} - u_{i}^{n}}{\Delta{t}} + u_{i}^{n}\frac{u_{i}^{n} - u_{i - 1}^{n}}{\Delta{x}} = 0  $$

This can be rearranged:

$$  u_{i}^{n + 1} = u_{i}^{n} - u_{i}^{n}\frac{\Delta{t}}{\Delta{x}}(u_{i}^{n} - u_{i - 1}^{n})  $$

```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams["figure.figsize"] = (16, 4)
```

```python
nx = 41

dx = 2 / (nx - 1)
n_steps = 25
dt = 0.025
c = 1
```

```python
u = np.ones(nx)
u[int(0.5 / dx) : int(1.0 // dx + 1)] = 2

u
```

```python
plt.plot(np.linspace(0, 2, nx), u)
```

```python
un = np.ones(nx)
for n in range(n_steps):
    
    un = u.copy()
    for i in range(1, nx):
        u[i] = un[i] - un[i] * dt / dx *(un[i] - un[i - 1])
```

```python
plt.plot(np.linspace(0, 2, nx), u)
```
